```python
import matplotlib.pyplot as plt
import networkx as nx

# Helper function to draw a graph in a subplot
def draw_topology(ax, G, pos, title):
    nx.draw(G, pos, ax=ax, with_labels=True, node_color='skyblue', node_size=1000, edge_color='gray')
    ax.set_title(title)
    ax.axis('off')

# Create figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Centralized
G1 = nx.Graph()
G1.add_node("Central")
for i in range(1, 5):
    G1.add_edge("Central", f"Node{i}")
pos1 = nx.spring_layout(G1)
draw_topology(axs[0, 0], G1, pos1, "Centralized")

# 2. Decentralized (Tree)
G2 = nx.DiGraph()
G2.add_edges_from([
    ("Root", "Node1"),
    ("Root", "Node2"),
    ("Node1", "Node3"),
    ("Node1", "Node4")
])
pos2 = nx.spring_layout(G2)
draw_topology(axs[0, 1], G2, pos2, "Decentralized (Tree)")

# 3. Decentralized (Parameter Server)
G3 = nx.Graph()
G3.add_node("ParamServer")
for i in range(1, 7):
    G3.add_edge("ParamServer", f"Worker{i}")
pos3 = nx.spring_layout(G3)
draw_topology(axs[1, 0], G3, pos3, "Decentralized (Parameter Server)")

# 4. Fully Distributed (Peer-to-Peer)
G4 = nx.complete_graph(6)
mapping = {i: f"Peer{i+1}" for i in range(6)}
G4 = nx.relabel_nodes(G4, mapping)
pos4 = nx.circular_layout(G4)
draw_topology(axs[1, 1], G4, pos4, "Fully Distributed (Peer-to-Peer)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define node properties
node_size = 1000
font_size = 14
worker_color = '#FFB6C1'     # Light pink
server_color = '#ADD8E6'     # Light blue

# Create directed graph
G3 = nx.DiGraph()

# Define parameter servers and workers
param_servers = [f'PS{i}' for i in range(1, 4)]
workers = [f'Worker{i}' for i in range(1, 7)]

# Add nodes
G3.add_nodes_from(param_servers)
G3.add_nodes_from(workers)

# Connect each worker to all parameter servers
for worker in workers:
    for ps in param_servers:
        G3.add_edge(worker, ps)

# Manual layout: workers at bottom, servers at top
pos = {}
for i, worker in enumerate(workers):
    pos[worker] = (i * 0.15, 0.1)
for i, ps in enumerate(param_servers):
    pos[ps] = (i * 0.3 + 0.15, 0.8)

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle('Parameter Server Architecture (Sharded, Shared Memory)', fontsize=18)

# Draw workers (circles)
nx.draw_networkx_nodes(G3, pos, nodelist=workers, node_color=worker_color,
                       node_size=node_size, node_shape='o', ax=ax)

# Draw parameter servers (squares)
nx.draw_networkx_nodes(G3, pos, nodelist=param_servers, node_color=server_color,
                       node_size=node_size, node_shape='s', ax=ax)

# Draw edges and labels
nx.draw_networkx_edges(G3, pos, ax=ax)
nx.draw_networkx_labels(G3, pos, font_size=font_size, ax=ax)

# Hide axes
ax.axis('off')
plt.tight_layout()
plt.show()
```
